# challenge question

```python
test_cats = image_test[image_test['label'] == 'cat']
test_dogs = image_test[image_test['label'] == 'dog']
test_birds = image_test[image_test['label'] == 'bird']
test_cars = image_test[image_test['label'] == 'automobile']
```

```python
def get_neighbors(nn_model, test_images):
    return nn_model.query(test_images, k=1)
```

```python
dog_neighbors_of_dogs = get_neighbors(dog_model, test_dogs)
get_images_from_ids(dog_neighbors_of_dogs)['image'].show()
```

```python
cat_neighbors_of_dogs = get_neighbors(cat_model, test_dogs)
get_images_from_ids(cat_neighbors_of_dogs)['image'].show()
```

```python
bird_neighbors_of_dogs = get_neighbors(bird_model, test_dogs)
get_images_from_ids(bird_neighbors_of_dogs)['image'].show()
```

```python
car_neighbors_of_dogs = get_neighbors(car_model, test_dogs)
get_images_from_ids(car_neighbors_of_dogs)['image'].show()
```

```python
dog_neighbors_of_dogs
```

```python
dog_distances = graphlab.SFrame(
        {'dog-dog': dog_neighbors_of_dogs['distance'],
         'dog-cat': cat_neighbors_of_dogs['distance'],
         'dog-bird': bird_neighbors_of_dogs['distance'],
         'dog-car': car_neighbors_of_dogs['distance']
        })
```

```python
dog_distances.head()
```

```python
dog_distances[0]['dog-cat']
```

```python
def is_dog_correct(row):
    dog_dog_distance = row['dog-dog']
    for distance in [row[k] for k in ['dog-cat', 'dog-bird', 'dog-car']]:
        if distance < dog_dog_distance:
            return 0
    return 1
```

```python
is_dog_correct(dog_distances[9])
```

```python
correct_dogs = dog_distances.apply(is_dog_correct)
len(correct_dogs)
```

```python
num_correct_dogs = correct_dogs.sum()
num_correct_dogs
```
